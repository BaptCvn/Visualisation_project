```python
import pandas as pd
from IPython.display import HTML
from IPython.display import Video
```

```python
data = pd.read_csv('Names_hints\dpt2020.csv', sep=';')
```

```python
data = data.drop(columns=["dpt"])
data = data[data["preusuel"] != "_preusuelS_RARES"]
data = data[data["annais"] != "XXXX"]
```

```python
data.head()
```

```python
import bar_chart_race as bcr


# Regrouper par année et prénom pour additionner les occurrences (au cas où)
grouped = data.groupby(['annais', 'preusuel'])['nombre'].sum().reset_index()

# Pivot pour avoir les prénoms en colonnes et les années en lignes
pivot = grouped.pivot(index='annais', columns='preusuel', values='nombre').fillna(0)

# Optionnel : garder uniquement les 100 prénoms les plus fréquents au total
# Cela rend l'animation plus fluide et la vidéo plus légère
top_preusuels = pivot.sum().sort_values(ascending=False).head(100).index
pivot = pivot[top_preusuels]

# Créer la vidéo avec bar_chart_race
bcr.bar_chart_race(
    df=pivot,
    filename='top_prenoms.mp4',
    orientation='h',
    sort='desc',
    n_bars=10,
    fixed_order=False,
    fixed_max=True,
    steps_per_period=1,
    interpolate_period=True,
    period_length=120,
    period_fmt='%Y',
    title='Top 10 des prénoms les plus donnés par année en France',
    bar_size=0.95,
    cmap='dark12',
    filter_column_colors=True
)
```
